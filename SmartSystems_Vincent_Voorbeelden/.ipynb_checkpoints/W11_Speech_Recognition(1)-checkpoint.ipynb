# Visualisation of Audio signals

```python
import numpy as np
import matplotlib.pyplot as plt
import os
import argparse
import warnings
from hmmlearn import hmm
from python_speech_features import mfcc, logfbank
from scipy.io import wavfile
from scipy.io.wavfile import write
import json
%matplotlib inline
```

```python
# reading the audio file
sampling_freq, signal = wavfile.read("random_sound.wav")
# display the params
print("Sampling Frequency: {} Hz".format(sampling_freq))
print("Signal Shape: {}".format(signal.shape))
print("Signal Duration: {} seconds".format(round(signal.shape[0]/ float(sampling_freq),2)))
```

```python
# Normalizing the signal
norm_signal = signal/ np.power(2,15)
```

```python
# Extracting some values for plotting
v_norm_signal = norm_signal[:75]
```

```python
# Building the time axis for the plot (in ms)
t_axis = 1000 * np.arange(0,len(v_norm_signal),1)/float(sampling_freq)
```

```python
# Signal plot
plt.plot(t_axis, v_norm_signal, color="blue")
plt.xlabel('Time - ms')
plt.ylabel('Magnitude')
plt.title('Normalized Raw Audio')
plt.show()
```

 # FFT of Signal

```python
len_v_norm_signal = len(v_norm_signal)
half_len_v_norm_signal = np.ceil((len_v_norm_signal + 1 )/2.0).astype(np.int)
print("Length of Norm Signal: {}".format(len_v_norm_signal))
print("Half length of Norm Signal: {}".format(half_len_v_norm_signal))
```

```python
# Apply Fourier to signal
v_freq_signal = np.fft.fft(norm_signal)
# Normalization
v_norm_freq_signal = abs(v_freq_signal[0 : half_len_v_norm_signal])/ len_v_norm_signal
# Take the square
v_norm_freq_signal **=2
# Extract length of frequency transformed signal
len_v_norm_freq_signal_fts=len(v_norm_freq_signal)
# Adjust signal for even and odd cases
if len_v_norm_freq_signal_fts % 2:
    v_norm_freq_signal[1:len_v_norm_freq_signal_fts] *= 2
else:
    v_norm_freq_signal[1:len_v_norm_freq_signal_fts-1] *= 2
# Extract power value in dB
v_norm_signal_power = 10 * np.log10(v_norm_freq_signal)
```

```python
# Build X-axis in kHz
f_axis = np.arange(0,half_len_v_norm_signal,1) * (sampling_freq / len_v_norm_signal) / 1000
```

```python
# Figure plot
plt.figure()
plt.plot(f_axis, v_norm_signal_power,color = 'blue')
plt.xlabel("Frequency in kHz")
plt.ylabel("Magnitude in dB")
plt.show()
```

# Generating Audio files

```python
# Output file name
file_name = "Audio_gen.wav"
# Signal parameters
duration = 3
sampling_frequency = 44100 #Hz
tone_frequency = 440
minimum_value = -4 * np.pi
maximum_value = 4 * np.pi
# Generate Audio signal
temp = np.linspace(minimum_value, maximum_value, duration * sampling_frequency)
signal = np.sin(2* np.pi * tone_frequency * temp)
```

```python
# Adding noise to the generated signal
noise = 0.3 * np.random.rand(duration * sampling_frequency)
signal += noise
```

```python
# Normalize and Scale the generated signal to 16-bit values
scaling_factor = np.power(2, 15) - 1
signal_normalized = signal / np.max(np.abs(signal))
signal_scaled = np.int16(signal_normalized * scaling_factor)
```

```python
# Save Generated Audio signal
write(file_name, sampling_frequency, signal_scaled)
```

```python
# Extracting some values for plotting
v_signal_scaled = signal_scaled[:75]
# Building the time axis for the plot (in ms)
t_axis = 1000 * np.arange(0,len(v_signal_scaled),1)/float(sampling_frequency)
# Signal plot
plt.plot(t_axis, v_signal_scaled, color="blue")
plt.xlabel('Time - ms')
plt.ylabel('Magnitude')
plt.title('Normalized Raw Audio')
plt.show()
```

# Generating Music Tones

```python
# Be sure to download the tone mapping file (json structure): http://www.phy.mtu.edu/~suits/notefreqs.html

# Tone input parameters
def tone_synth(frequency, duration, amplitude=1.0, sampling_frequency = 44100):
    # t-axis
    t_axis = np.linspace(0, duration, int(duration* sampling_frequency))
    # Audio signal
    signal = amplitude * np.sin(2 * np.pi * frequency * t_axis)
    
    return signal.astype(np.int16)
```

```python
# filename definitions
file_single_tone = "single_tone.wav"
file_multi_tone = "multi_tone.wav"
# load tone mapping file
map_file = "tone_mapping.json"
with open(map_file, 'r') as f:
    tone_map = json.loads(f.read())
```

```python
# Generate F tone with duration of 3 seconds
t_name = 'F'
duration = 3 # seconds
amplitude = 12000
sampling_frequency = 44100 # Hz
tone_frequencies = tone_map[t_name]
synth_tone = tone_synth(tone_frequencies,duration,amplitude,sampling_frequency)
write(file_single_tone,sampling_frequency,np.int16(synth_tone))
```

```python
# Multi tone
t_sequence = [('G',0.4),('D',0.5),('F',0.3),('C',0.6),('A',0.4)]
m_tone_signal = np.array([])
for item in t_sequence:
    tone_name = item[0]
    freq= tone_map[tone_name]
    duration = item[1]
    synth_tone = tone_synth(freq,duration,amplitude,sampling_frequency)
    m_tone_signal = np.append(m_tone_signal,synth_tone,axis=0)
write(file_multi_tone,sampling_frequency, np.int16(m_tone_signal))
```

```python
import os
import argparse 
import warnings

import numpy as np
from scipy.io import wavfile 

from hmmlearn import hmm
from python_speech_features import mfcc

# Define a function to parse the input arguments
def build_arg_parser():
    parser = argparse.ArgumentParser(description='Trains the HMM-based speech \
            recognition system')
    parser.add_argument("--input-folder", dest="input_folder", required=True,
            help="Input folder containing the audio files for training")
    return parser

# Define a class to train the HMM 
class ModelHMM(object):
    def __init__(self, num_components=4, num_iter=1000):
        self.n_components = num_components
        self.n_iter = num_iter

        self.cov_type = 'diag' 
        self.model_name = 'GaussianHMM' 

        self.models = []

        self.model = hmm.GaussianHMM(n_components=self.n_components, 
                covariance_type=self.cov_type, n_iter=self.n_iter)

    # 'training_data' is a 2D numpy array where each row is 13-dimensional
    def train(self, training_data):
        np.seterr(all='ignore')
        cur_model = self.model.fit(training_data)
        self.models.append(cur_model)

    # Run the HMM model for inference on input data
    def compute_score(self, input_data):
        return self.model.score(input_data)

# Define a function to build a model for each word
def build_models(input_folder):
    # Initialize the variable to store all the models
    speech_models = []

    # Parse the input directory
    for dirname in os.listdir(input_folder):
        # Get the name of the subfolder 
        subfolder = os.path.join(input_folder, dirname)

        if not os.path.isdir(subfolder): 
            continue

        # Extract the label
        label = subfolder[subfolder.rfind('/') + 1:]

        # Initialize the variables
        X = np.array([])

        # Create a list of files to be used for training
        # We will leave one file per folder for testing
        training_files = [x for x in os.listdir(subfolder) if x.endswith('.wav')][:-1]

        # Iterate through the training files and build the models
        for filename in training_files: 
            # Extract the current filepath
            filepath = os.path.join(subfolder, filename)

            # Read the audio signal from the input file
            sampling_freq, signal = wavfile.read(filepath)
            
            # Extract the MFCC features
            with warnings.catch_warnings():
                warnings.simplefilter('ignore')
                features_mfcc = mfcc(signal, sampling_freq)

            # Append to the variable X
            if len(X) == 0:
                X = features_mfcc
            else:
                X = np.append(X, features_mfcc, axis=0)
            
        # Create the HMM model
        model = ModelHMM()

        # Train the HMM
        model.train(X)

        # Save the model for the current word
        speech_models.append((model, label))

        # Reset the variable
        model = None

    return speech_models

# Define a function to run tests on input files
def run_tests(test_files):
    # Classify input data
    for test_file in test_files:
        # Read input file
        sampling_freq, signal = wavfile.read(test_file)

        # Extract MFCC features
        with warnings.catch_warnings():
            warnings.simplefilter('ignore')
            features_mfcc = mfcc(signal, sampling_freq)

        # Define variables
        max_score = -float('inf') 
        output_label = None 

        # Run the current feature vector through all the HMM
        # models and pick the one with the highest score
        for item in speech_models:
            model, label = item
            score = model.compute_score(features_mfcc)
            if score > max_score:
                max_score = score
                predicted_label = label

        # Print the predicted output 
        start_index = test_file.find('/') + 1
        end_index = test_file.rfind('/')
        original_label = test_file[start_index:end_index]
        print('\nOriginal: ', original_label) 
        print('Predicted:', predicted_label)

if __name__=='__main__':
    # args = build_arg_parser().parse_args()
    # input_folder = args.input_folder
    input_folder = "audio"
    # Build an HMM model for each word
    speech_models = build_models(input_folder)

    # Test files -- the 15th file in each subfolder 
    test_files = []
    for root, dirs, files in os.walk(input_folder):
        for filename in (x for x in files if '15' in x):
            filepath = os.path.join(root, filename)
            test_files.append(filepath)

    run_tests(test_files)
```
