# Basic Neural Network with TensorFlow

```python
import numpy as np # linear algebra
np.random.seed(101)
tf.set_random_seed(101)
```

```python
rand_a = np.random.uniform(0,100,(5,5))
rand_b = np.random.uniform(0,100,(5,1))

a = tf.placeholder(tf.float32)
b = tf.placeholder(tf.float32)
add_op = a+b
mult_op = a * b
```

```python
with tf.Session() as sess:
    s = sess.run(add_op, feed_dict={a:rand_a, b:rand_b})
    print(s)
    
    m = sess.run(mult_op, feed_dict={a:rand_a, b: rand_b})
    print(m)
```

```python
# we are going to use 10 features
n_features = 10
n_dense_neurons = 3

# Placeholder for x
x = tf.placeholder(tf.float32,(None,n_features))

# Variables for w and b
b = tf.Variable(tf.zeros([n_dense_neurons]))
W = tf.Variable(tf.random_normal([n_features,n_dense_neurons]))
```

```python
# Activation function
xW = tf.matmul(x,W)
z = tf.add(xW,b)
# tf.nn.relu() or tf.tanh()
a = tf.sigmoid(z)

# Variable initialization
init = tf.global_variables_initializer()
with tf.Session() as sess:
    sess.run(init)
    print(sess.run(z,feed_dict={x: np.random.random([1,n_features])}))
    layer_out = sess.run(a,feed_dict={x : np.random.random([1, n_features])})
    
print(layer_out)
```

## Full Neural Network example
### y = mx + b

```python
g= tf.Graph()
np.linspace(1,10,10)
```

```python
# Artificial data
xdata = np.linspace(0,10,10) + np.random.uniform(-1.5,1.5,10)
print(xdata)
ydata = np.linspace(0,10,10) + np.random.uniform(-1.5,1.5,10)
print(ydata)
```

```python
m = tf.Variable(0.39)
b = tf.Variable(0.2)
```

```python
# Cost function or Residual
error = 0
for x,y  in zip(xdata, ydata):
    y_pred = m*x + b
    error += (y-y_pred)**2
print(error)
```

```python
# Optimizer
optimizer = tf.train.GradientDescentOptimizer(learning_rate=0.001)
train = optimizer.minimize(error)
train
```

```python
init  = tf.global_variables_initializer()
with tf.Session() as sess:
    sess.run(init)
    epochs = 1
    for i in range(epochs):
        sess.run(train)
    
    # Fetch back Results
    final_slope, final_intercept = sess.run([m,b])
```

```python
print(final_slope)
print(final_intercept)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt

x_test = np.linspace(-1,11,10)
y_pred_plot = final_slope*x_test + final_intercept
plt.plot(x_test,y_pred_plot,'r')
plt.plot(xdata,ydata,'*')
plt.show()
```

## Polynomial Lineair Regression example 2

```python
# ISess.close() => Interactive session from above uncomment this if you get errors for interactive session is running,...
# sess.close() => Same, sometimes I use the name sess also for my interactive sessions
%matplotlib notebook
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
```

```python
%matplotlib notebook
%matplotlib notebook
# %matplotlib inline
plt.ion()
n_observations=100
fig,ax = plt.subplots(1,1)
xs = np.linspace(-3,3,n_observations)
ys = np.sin(xs) # + np.random.uniform(-0.5,0.5,n_observations)
ax.scatter(xs,ys)
fig.show()
plt.draw()

# Create Placeholders for variable X and Y
X = tf.placeholder(tf.float32)
Y = tf.placeholder(tf.float32)

# Create variable Y_pred

Y_pred = tf.Variable(tf.random_normal([1]),name='bias')
for pow_i in range(1,5):
    W = tf.Variable(tf.random_normal([1]),name='weight_%d' % pow_i)
    Y_pred = tf.add(tf.multiply(tf.pow(X,pow_i),W),Y_pred)
    
# Polynomial function    
# Y_pred = w4X**4 + w3X**3 + w2X**2 + w1X**1

# cost function to reduce cost
cost = tf.reduce_sum(tf.pow(Y_pred - Y,2))/ (n_observations - 1)



# Setting the learning rate and optimizer
learning_rate = 0.01
optimizer = tf.train.GradientDescentOptimizer(learning_rate).minimize(cost)

n_epochs = 1000
with tf.Session() as sess:
    sess.run(tf.global_variables_initializer())
    prev_training_cost = 0.0
    for epoch_i in range(n_epochs):
        for(x,y) in zip(xs,ys):
            sess.run(optimizer,feed_dict={X: x, Y: y})
            
        training_cost = sess.run(cost,feed_dict={X: xs, Y: ys})
        print (training_cost)
        
        if epoch_i % 100 == 0:
            ax.plot(xs, Y_pred.eval(feed_dict={X: xs}, session =sess),'k',alpha=epoch_i/n_epochs)
            fig.show()
            plt.draw()
        if np.abs(prev_training_cost - training_cost) < 0.000001:
            break
        prev_training_cost = training_cost
ax.set_ylim([-3,3])
fig.show()
```
